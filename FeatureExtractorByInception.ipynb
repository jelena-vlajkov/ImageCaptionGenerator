```python
import glob
import numpy as np
import pickle
import pandas as pd
from keras.applications.inception_v3 import InceptionV3
from keras.preprocessing import image
```

```python
DATASET_PATH = "drive/MyDrive/flickr8/captions.txt"
IMAGES_PATH = "drive/MyDrive/flickr8/Images/"
```

```python
model = InceptionV3(weights='imagenet')
```

```python
from keras.models import Model

new_input = model.input
hidden_layer = model.layers[-2].output

feature_extractor = Model(new_input, hidden_layer)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv(DATASET_PATH)
```

```python
def preprocess_input(x):
    x /= 255.
    x -= 0.5
    x *= 2.
    return x
```

```python
def preprocess(image_path):
    img = image.load_img(image_path, target_size=(299, 299))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)

    x = preprocess_input(x)
    return x
```

```python
def encode(image):
    image = preprocess(IMAGES_PATH + image)
    temp_enc = feature_extractor.predict(image)
    temp_enc = np.reshape(temp_enc, temp_enc.shape[1])
    return temp_enc
```

```python
encoded_dict = {}

for index, row in df.iterrows():
  try:
    print(row["image"])
    if row["image"] not in encoded_dict:
      encoded = encode(row["image"])
      encoded_dict[row["image"]] = encoded
  except Exception as e:
    print(e)
    continue
```
